# Can network fit random labels?

## 采用的模型

| Model                                                        | Activation |
| :----------------------------------------------------------- | :-------: |
| model： `Conv(16, (5x5))->Pool(2)->Conv(32, (5x5))->Conv(32, (5x5))->Conv(64, (5x5))->Pool(2)->Conv(10, (1x1))->Pool(7)` |   BN+ReLU    |

说明：
1. 这部分为了加快收敛的速度，采用了更大的网络结果来拟合MNIST，以及加入BN层（这点和之前采用的三个MNIST模型不同）
2. Adam优化加了amsgrad=True来加快收敛

```python
import sys
sys.path.insert(0, '..')
import os
import torch
from argparse import Namespace
from torch import optim
from model import cross_entropy_loss, accuracy
from model import deep_mnist_bn
from dataset import mnist_train_loader, mnist_validate_loader
from solver import HW1Solver
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# 设置参数 --- 你可以令visdom=True, 然后在http://localhost:8097上看训练过程
root = os.path.join(os.path.expanduser('~'), 'data')
config = {'batch_size': 400, 'epochs': 500, 'resume': '', 'verbosity': 1, 'use_cuda': True,
          'lr': 1e-3, 'save_dir': '../pretrained/1_3_1', 'save_freq': 50, 'save_grad': False,
          'data_dir': root, 'dataset': 'mnist', 'valid': True, 'val_step': 1,
          'visdom': False, 'visdom_iter': True, 'visdom_fit': False}
config = Namespace(**config)
```

```python
# 训练阶段  --- 可能需要花费几十分钟, 请耐心等待～
# ---友情提示: 你也可以直接使用提供的已训练好的模型, 那么你就别运行这部分了 ---
model = deep_mnist_bn()
optimizer = optim.Adam(model.parameters(), config.lr, amsgrad=True)
loss = cross_entropy_loss
metrics = [accuracy]
train_loader = mnist_train_loader(root, config.batch_size, target_random=True)
val_loader = mnist_validate_loader(root, config.batch_size)
solver = HW1Solver(model, optimizer, loss, metrics, train_loader, val_loader, config)
solver.train()
```

```python
# 可视化
f, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 8))
ax1.set_title('mnist loss (train v.s val)')
ax2.set_title('mnist accuracy (train v.s val)')
checkpoint = torch.load('{}/deep_mnist_bn_epoch{}.pth.tar'.format(config.save_dir, config.epochs))
logger = checkpoint['logger']
x = [entry['epoch'] for _, entry in logger.entries.items()]
loss_train = [entry['loss'] for _, entry in logger.entries.items()]
loss_val = [entry['val_loss'] for _, entry in logger.entries.items()]
acc_train = [entry['accuracy'] for _, entry in logger.entries.items()]
acc_val = [entry['val_accuracy'] for _, entry in logger.entries.items()]
ax1.plot(x, loss_train, 'r', label='train')
ax1.plot(x, loss_val, 'b', label='val')
ax1.legend(loc='best')
ax2.plot(x, acc_train, 'r', label='train')
ax2.plot(x, acc_val, 'b', label='val')
ax2.legend(loc='best')
plt.tight_layout()
plt.show()
```

### 结果分析

1. 可以发现即使是random label，只要网络参数足够多，训练次数足够，也能够在训练集上取得非常高的精度！其实还有个特点，testing loss其实反而在增加
2. 但是在"正常的label"的测试效果基本等同于随机乱猜（10%---10个结果随机猜一个）
3. 虽然网络足够"大"能够拟合random labels的情况，但付出的代价是需要更多的参数+更多的迭代次数（个人hw1_1里面的网络很难收敛）
