# Observe Gradient Norm During Training

## 采用的模型

① Simulate采用的模型

| Model                                     | Parameter |
| :---------------------------------------- | :-------: |
| deep model：  `1->5->10->10->10->10->10->5->1` |    571    |

> 注：上述模型采用Tanh作为激活函数
>
> 数据域：$[0,1,0.001]$---总共1000个数
>
> 训练采用的参数：`batch_size=250, optimizer=Adam(lr=1e-3), epoch=2000 `

② MNIST采用的模型

| Model                                                        | Activation |
| :----------------------------------------------------------- | :--------: |
| model：  `Conv(16, (3x3))->Pool(2)->Conv(16, (3x3))->Conv(16, (3x3))->Conv(16, (3x3))->Pool(2)->Conv(10, (1x1))->Pool(7)` |    ReLU    |

> 训练采用的参数：`batch_size=250, optimizer=Adam(lr=1e-3), epoch=80 `

```python
import sys
sys.path.insert(0, '..')
import os
import torch
import numpy as np
from argparse import Namespace
from torch import optim
from model import mse_loss, cross_entropy_loss
from model import deep_simulate, deep_mnist
from solver import HW1Solver
from dataset import simulate_dataloader, mnist_train_loader
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# 设置参数 --- 你可以令visdom=True, 然后在http://localhost:8097上看训练过程
root = os.path.join(os.path.expanduser('~'), 'data')
config = {'batch_size': 250, 'epochs': 2000, 'resume': '', 'verbosity': 1, 'save_dir': '../pretrained/1_2_2',
          'data_dir': root, 'use_cuda': False, 'dataset': 'simulate', 'save_freq': 500,
          'save_grad': True, 'lr': 1e-3, 'valid': False, 'val_step': 200, 'target_func': 'sin',
          'visdom': False, 'visdom_iter': False, 'visdom_fit': False}
config = Namespace(**config)
```

## 训练$sin(5\pi x)/(5\pi x)$

```python
# 训练过程 --- 可能需要花费几分钟, 请耐心等待～
# ---友情提示: 你也可以直接使用提供的已训练好的模型, 那么你就别运行这部分了 ---
model = deep_simulate()
optimizer = optim.Adam(model.parameters(), config.lr)
loss = mse_loss
metrics = []
train_loader = simulate_dataloader(config.target_func, config.batch_size)
solver = HW1Solver(model, optimizer, loss, metrics, train_loader, None, config)
solver.train()
```

```python
# 可视化结果
model_name = 'deep_simulate'
f, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 8))
ax1.set_title(model_name + ' loss')
ax2.set_title(model_name + ' loss(log)')
checkpoint = torch.load('{}/{}_epoch{}.pth.tar'.format(config.save_dir, model_name, config.epochs))
logger = checkpoint['logger']
epoch = [entry['epoch'] for _, entry in logger.entries.items()]
loss = [entry['loss'] for _, entry in logger.entries.items()]
grad_norm = [entry['grad_norm'] for _, entry in logger.entries.items()]
ax1.plot(epoch, grad_norm, 'b', label='gradnorm')
ax1.plot(epoch, loss, 'r', label='loss')
ax1.legend(loc="best")
ax2.semilogy(epoch, grad_norm, 'b', label='gradnorm')
ax2.semilogy(epoch, loss, 'r', label='loss')
ax2.legend(loc="best")
plt.tight_layout()
plt.show()
```

### 结果分析

1. 说实话，我并没有看出gradient norm与loss之间的关系，但是可以发现loss较低时候，gradient norm的值其实蛮小的
2. 但gradient norm的值始终没有等于0，虽然我们总是是网络收敛到local minima，但实际上如果根据定义而言，gradient norm应该等于0才对

### 训练MNIST

```python
# 一些参数需要调整
config = vars(config)
config.update({'use_cuda': True, 'epochs': 80, 'save_freq': 20, 'dataset': 'mnist'})
config = Namespace(**config)
```

```python
# 训练过程 --- 可能需要花费几分钟, 请耐心等待～
# ---友情提示: 你也可以直接使用提供的已训练好的模型, 那么你就别运行这部分了 ---
model = deep_mnist()
optimizer = optim.Adam(model.parameters(), config.lr)
loss = cross_entropy_loss
metrics = []
train_loader = mnist_train_loader(config.data_dir, config.batch_size)
solver = HW1Solver(model, optimizer, loss, metrics, train_loader, None, config)
solver.train()
```

```python
# 可视化结果
model_name = 'deep_mnist'
f, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 8))
ax1.set_title(model_name + ' loss')
ax2.set_title(model_name + ' log(loss)')
checkpoint = torch.load('{}/{}_epoch{}.pth.tar'.format(config.save_dir, model_name, config.epochs))
logger = checkpoint['logger']
epoch = [entry['epoch'] for _, entry in logger.entries.items()]
loss = [entry['loss'] for _, entry in logger.entries.items()]
grad_norm = [entry['grad_norm'] for _, entry in logger.entries.items()]
ax1.plot(epoch, grad_norm, 'b', label='gradnorm')
ax1.plot(epoch, loss, 'r', label='loss')
ax1.legend(loc="best")
ax2.semilogy(epoch, grad_norm, 'b', label='gradnorm')
ax2.semilogy(epoch, loss, 'r', label='loss')
ax2.legend(loc="best")
plt.tight_layout()
plt.show()
```

### 结果分析

1. 其实和仿真实验类似，我们能看到gradient norm降到较小的值，但gradient norm始终不等于0
