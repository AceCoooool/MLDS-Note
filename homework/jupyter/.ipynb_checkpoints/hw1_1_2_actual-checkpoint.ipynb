# Train on Actual Tasks

## 采用的模型

① MNIST采用的模型

| Model                                                        | Parameter |
| :----------------------------------------------------------- | :-------: |
| shallow model：  `Conv(208, (5x5))->Pool(4)->Conv(10, (1x1))->Pool(7)` |   7290    |
| middle model：  `Conv(24, (3x3))->Pool(2)->Conv(31, (3x3))->Pool(2)->Conv(10, (1x1))->Pool(7)` |   7232    |
| deep model：  `Conv(16, (3x3))->Pool(2)->Conv(16, (3x3))->Conv(16, (3x3))->Conv(16, (3x3))->Pool(2)->Conv(10, (1x1))->Pool(7)` |   7226    |

> 注：上述除最后一个卷积之外，每个卷积之后加ReLU激活函数
>
> 训练采用的参数：`batch_size=400, optimizer=Adam(lr=1e-3), epoch=100 `

② CIFAR采用的模型

| Model                                                        | Parameter |
| :----------------------------------------------------------- | :-------: |
| shallow model：  `Conv(128, (3x3))->Pool(2)->Conv(80, (3x3))->Pool(2)->Conv(10, (1x1))->Pool(8)` |   96842   |
| middle model：  `Conv(32, (3x3))->Pool(2)->Conv(64, (3x3))->Conv(128, (3x3))->Pool(2)->Conv(10, (1x1))->Pool(8)` |   94762   |
| deep model：  `Conv(16, (3x3))->Pool(2)->Conv(32, (3x3))->Conv(64, (3x3))->Conv(64, (3x3))->Pool(2)->Conv(38, (3x3))->Conv(32, (3x3))->Conv(10, (1x1))->Pool(8)` |   93990   |

> 注：上述除最后一个卷积之外，每个卷积之后 加BatchNormalization和ReLU激活函数
>
> 训练采用的参数：`batch_size=400, optimizer=Adam(lr=1e-3), epoch=100 `

```python
import sys
sys.path.insert(0, '..')
import os
import torch
from argparse import Namespace
from torch import optim
from model import cross_entropy_loss, accuracy
from model import deep_mnist, shallow_mnist, middle_mnist
from model import deep_cifar, shallow_cifar, middle_cifar
from dataset import mnist_train_loader, mnist_validate_loader
from dataset import cifar_train_loader, cifar_validate_loader
from solver import HW1Solver
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# 设置参数 --- 你可以令visdom=True, 然后在http://localhost:8097上看训练过程
root = os.path.join(os.path.expanduser('~'), 'data')
config = {'batch_size': 400, 'epochs': 100, 'resume': '', 'verbosity': 1, 'use_cuda': True,
          'lr': 1e-3, 'save_dir': '../pretrained/1_1_2', 'save_freq': 20, 'save_grad': False,
          'data_dir': root, 'valid': True, 'val_step': 1,
          'visdom': False, 'visdom_iter': True, 'visdom_fit': False}
```

## 训练MNIST

```python
dataset = 'mnist'
config.update({'dataset': dataset})
model_list = ['shallow_' + dataset, 'middle_' + dataset, 'deep_' + dataset]
config = Namespace(**config)
```

```python
# 训练MNIST： 这部分时间可能达到几十分钟; 
# ---友情提示: 你也可以直接使用提供的已训练好的模型, 那么你就别运行这部分了 ---
for name in model_list:
    model = eval(name)()
    optimizer = optim.Adam(model.parameters(), config.lr)
    loss = cross_entropy_loss
    metrics = [accuracy]
    train_loader = eval(dataset + '_train_loader')(root, config.batch_size)
    val_loader = eval(dataset + '_validate_loader')(root, config.batch_size)
    solver = HW1Solver(model, optimizer, loss, metrics, train_loader, val_loader, config)
    solver.train()
```

```python
# 可视化结果 
color_list = ['r', 'g', 'b']
f, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 8))
ax1.set_title(dataset + ' loss')
ax2.set_title(dataset + ' accuracy')
for arch, color in zip(model_list, color_list):
    checkpoint = torch.load('{}/{}_epoch{}.pth.tar'.format(config.save_dir, arch, config.epochs))
    logger = checkpoint['logger']
    epoch = [entry['epoch'] for _, entry in logger.entries.items()]
    loss = [entry['loss'] for _, entry in logger.entries.items()]
    accuracy = [entry['val_metrics'][0] for _, entry in logger.entries.items()]
    ax1.semilogy(epoch, loss, color, label=arch)
    ax1.legend(loc="best")
    ax2.plot(epoch, accuracy, color, label=arch)
    ax2.legend(loc="best")
plt.tight_layout()
plt.show()
```

### 结果分析

1. 可以很明显的发现Deep模型收敛更快，且结果更佳！

```python
# 训练CIFAR10： 这部分时间可能达到几十分钟; 
# ---友情提示: 你也可以直接使用提供的已训练好的模型, 那么你就别运行这部分了 ---
for name in model_list:
    model = eval(name)()
    optimizer = optim.Adam(model.parameters(), config.lr)
    loss = cross_entropy_loss
    metrics = [accuracy]
    train_loader = eval(dataset + '_train_loader')(root, config.batch_size)
    val_loader = eval(dataset + '_validate_loader')(root, config.batch_size)
    solver = HW1Solver(model, optimizer, loss, metrics, train_loader, val_loader, config)
    solver.train()
```

```python
# 可视化结果
f, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 8))
ax1.set_title(dataset + ' loss')
ax2.set_title(dataset + ' accuracy')
for arch, color in zip(model_list, color_list):
    checkpoint = checkpoint = torch.load('{}/{}_epoch{}.pth.tar'.format(config.save_dir, arch, config.epochs))
    logger = checkpoint['logger']
    epoch = [entry['epoch'] for _, entry in logger.entries.items()]
    loss = [entry['loss'] for _, entry in logger.entries.items()]
    accuracy = [entry['val_metrics'][0] for _, entry in logger.entries.items()]
    ax1.semilogy(epoch, loss, color, label=arch)
    ax1.legend(loc="best")
    ax2.plot(epoch, accuracy, color, label=arch)
    ax2.legend(loc="best")
plt.tight_layout()
plt.show()
```
